**Task # 1**

A. Download the dataset and explore how to merge these dataset.

B. Combine dataset Lab2 D1A with Lab2 D1B in such a way that it doesn’t contain any duplicate column. The resultant dataset consist of these columns and the final shape will be.

C. Combine dataset Lab2 D1A with Lab2 D1C using merge method to extract similar records in a new Dataframe “comboAC” having 4221333 records.


```python
import pandas as pd

df1 = pd.read_csv('/content/Lab2 D1A.csv')
df2 = pd.read_csv('/content/Lab2 D1B.csv')
df3 = pd.read_csv('/content/Lab2 D1C.csv')

#  B: Merge Lab2_D1A with Lab2_D1B, avoiding duplicate columns-->final shape
df1_columns = set(df1.columns)
df2_columns = set(df2.columns)
common_columns = list(df1_columns & df2_columns)
merged_df = pd.merge(df1, df2, on=common_columns, how='inner')
print("Merged DataFrame from Lab2_D1A and Lab2_D1B:")
print(merged_df.head())
print("Shape:", merged_df.shape)

# C: Merge Lab2_D1A with Lab2_D1C to create comboAC
comboAC = pd.merge(df1, df3, how='inner')
print("\ncomboAC DataFrame:")
print(comboAC.head())
print("Shape:", comboAC.shape)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = '/content/train.csv'
df = pd.read_csv(url)
print("First few rows of the dataset:")
df.head()

# C. Calculate the correlation between the specified variables
variables_of_interest = ['var3', 'var38', 'var15', 'imp_op_var39_comer_ult1']
correlation_matrix = df[variables_of_interest].corr()
print("\nCorrelation matrix:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# D Check linearity by plotting pairwise relationships
sns.pairplot(df[variables_of_interest])
plt.show()

# E.Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values[missing_values > 0])

# G. Handle missing data by imputing with the average (mean)
df.fillna(df.mean(), inplace=True)
target_variable = 'TARGET'
unique_categories = df[target_variable].unique()

print(f"\nUnique categories in the target variable '{target_variable}':")
print(unique_categories)

# H. Check if the dataset is balanced AND Handle class imbalance using upsampling
target_counts = df[target_variable].value_counts()
print(f"\nClass distribution in target variable '{target_variable}':")
print(target_counts)
max_count = target_counts.max()
df_balanced = df.copy()

for category in unique_categories:
    category_data = df[df[target_variable] == category]
    upsampled_data = category_data.sample(max_count, replace=True, random_state=42)
    df_balanced = pd.concat([df_balanced, upsampled_data])

df_balanced = df_balanced.drop_duplicates()

# I. Summary statistics for the selected features,  Pearson Correlation and feature selection,
print("\nSummary statistics for the selected features:")
summary = df.describe()
print(summary)
correlation_matrix = df.corr()
high_corr_features = correlation_matrix.index[abs(correlation_matrix['TARGET']) >= 0.65]
print("\nFeatures with high correlation (>= 0.65) with TARGET:")
print(high_corr_features)

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

X = df.drop(columns=[target_variable])
y = df[target_variable]
logreg = LogisticRegression(max_iter=1000)
rfe = RFE(logreg, n_features_to_select=10)
rfe = rfe.fit(X, y)
selected_features = X.columns[rfe.support_]
print("\nSelected features using RFE:")
print(selected_features)
df.to_csv('/content/train.csv', index=False)
print("\nProcessed dataset saved as 'processed_dataset.csv'.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, accuracy_score
from imblearn.under_sampling import RandomUnderSampler

file_path = '/content/Customer_support_data.csv'
df = pd.read_csv(file_path)

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

df[df.select_dtypes(include=[np.number]).columns] = num_imputer.fit_transform(df.select_dtypes(include=[np.number]))
df[df.select_dtypes(include=[object]).columns] = cat_imputer.fit_transform(df.select_dtypes(include=[object]))
df.drop_duplicates(inplace=True)
print(df.describe())

sns.countplot(x='CSAT Score', data=df)
plt.title('Distribution of CSAT Score')
plt.show()

sns.countplot(x='CSAT Score', data=df)
plt.title('CSAT Score Distribution')
plt.show()

X = df.drop('CSAT Score', axis=1)
y = df['CSAT Score']
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
X_res_encoded = ohe.fit_transform(X_res)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_res_encoded, y_res)
sfm = SelectFromModel(clf, threshold=0.01)
sfm.fit(X_res_encoded, y_res)
X_important = sfm.transform(X_res_encoded)
if X_important.shape[1] == 0:
    print("Warning: No features were selected. Consider adjusting the threshold.")
else:
    X_train, X_test, y_train, y_test = train_test_split(X_important, y_res, test_size=0.2, random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(classification_report(y_test, y_pred))
    print('Accuracy:', accuracy_score(y_test, y_pred))

duplicate_rows = df[df.duplicated()]
print(f'Duplicate Rows Count: {duplicate_rows.shape[0]}')

df_no_duplicates = df.drop_duplicates(keep='first')

df_no_duplicates.to_csv
```
