```python
# A. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# B. Load the dataset from the given URL
url = "https://raw.githubusercontent.com/Engr-mehak/ML_Dataset/main/cancer%20patient%20data%20sets.csv"
df = pd.read_csv(url)

# C. Find unique categories in target variable
print(df['Level'].unique())

# D. Explore dataset information
print(df.info())
print(df.head(50))
print(df.tail(30))
print(df.memory_usage())
print(df.shape)

# E. Find numbers of columns and total number of samples
print(df.shape)

# F. Find number of records for each unique category of target variable
print(df['Level'].value_counts())
sns.countplot(x='Level', data=df)
plt.show()

# G. Extract sample from 550th to 900th sample
sample1 = df.loc[550:900, 'Gender':'Coughing of Blood']
sample1.to_csv('sample1.csv', index=False)
print(sample1.describe())

# H. Extract sample of last 20 records
sample2 = df.iloc[-20:, 5:10]
sample2.to_csv('sample2.csv', index=False)
print(sample2.describe())

# I. Extract sample of specific records
sample3 = df.loc[[110, 220, 360, 440, 656, 778, 202, 889], :]
print(sample3['Alcohol use'].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

url = 'https://raw.githubusercontent.com/Engr-mehak/ML_Dataset/main/cancer%20patient%20data%20sets.csv'
df = pd.read_csv(url)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# A. Graphical Visualization

# B. Display a correlation graph
plt.figure(figsize=(12, 10))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# C. Display the distribution of Age and Frequent Cold
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].hist(df['Age'], bins=20, color='skyblue')
ax[0].set_title('Age Distribution')
ax[1].hist(df['Frequent Cold'], bins=20, color='lightgreen')
ax[1].set_title('Frequent Cold Distribution')
plt.show()

# D. Find out outliers in Obesity feature by using Graphical Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Obesity'])
plt.title('Obesity Outliers')
plt.show()
```
